```python
import requests
import json
import pandas as pd
import re
```

```python
def weather_forecast(cities):

  API_key = '197ff6b262aa1061aa253928df03ed33'

  weather_dict = {'city': [],
                'country': [],
                'latitude': [],
                'longitude': [],
                'temp': [],
                'temp_min': [],
                'temp_max': [],
                'date_of_forecast': [],
                'weather_overview': [], 
                'weather_description': [], 
                'feels_like': [], 
                'pressure': [],
                'humidity': [], 
                'wind_speed': [],
                'wind_deg': [], 
                'clouds': []
                }

  for city in cities:
    url = (f"http://api.openweathermap.org/data/2.5/forecast?q={city}&appid={API_key}&units=metric")
    response = requests.get(url)
    json = response.json()
    weather_df = pd.json_normalize(json["list"])

    for i in range(0,len(weather_df),8):
      weather_dict['city'].append(json['city']['name'])
      weather_dict['country'].append(json['city']['country'])
      weather_dict['latitude'].append(json['city']['coord'].get('lat'))
      weather_dict['longitude'].append(json['city']['coord'].get('lon'))
      weather_dict['temp'].append(json["list"][i]['main']['temp'])
      weather_dict['temp_min'].append(json["list"][i]['main']['temp_min'])
      weather_dict['temp_max'].append(json["list"][i]['main']['temp_max'])
      weather_dict['date_of_forecast'].append(json["list"][i]['dt_txt'])
      weather_dict["weather_overview"].append(json["list"][i]['weather'][0]["main"])
      weather_dict["weather_description"].append(json["list"][i]['weather'][0]["description"])
      weather_dict["feels_like"].append(json["list"][i]['main']["feels_like"])
      weather_dict["pressure"].append(json["list"][i]['main']["pressure"])
      weather_dict["humidity"].append(json["list"][i]['main']["humidity"])
      weather_dict["wind_speed"].append(json["list"][i]["wind"]["speed"])
      weather_dict["wind_deg"].append(json["list"][i]["wind"]["deg"])
      weather_dict["clouds"].append(json["list"][i]["clouds"]["all"])

  return pd.DataFrame(weather_dict)
```
