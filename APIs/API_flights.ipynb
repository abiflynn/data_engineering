```python
def flight_arrivals(icao,date_from,date_to):

  list_for_df = []
  for x in icao:
    url = f"https://aerodatabox.p.rapidapi.com/flights/airports/icao/{icao}/{date_from}/{date_to}"
    querystring = {"withLeg":"true","withCancelled":"true","withCodeshared":"true","withCargo":"true","withPrivate":"true","withLocation":"false"}
    headers = {
                  "X-RapidAPI-Key": "00663ce840msh1d6492d6b0bcbc8p1d356ajsnfcd4007adece",
                  "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
              }

    flights = requests.request("GET", url, headers=headers, params=querystring)
    flights = flights.json()

    for i in flights["arrivals"]:
        flights_dict = {}
        
        flights_dict['flight_number'] = i.get('number', None)
        flights_dict['flight_status'] = i.get("status", None)
        flights_dict['airline'] = i['airline'].get('name', None)
        flights_dict['cargo'] = i.get('isCargo', None)
        flights_dict['departure_iata'] = i["departure"]["airport"].get("iata", None)
        flights_dict['departure_city'] = i['departure']['airport'].get('name', None)
        flights_dict['local_departure_time'] = i['departure'].get('scheduledTimeLocal', None)
        flights_dict['utc_departure_time'] = i['departure'].get('scheduledTimeUtc', None)
        flights_dict["arrival_icao"] = icao
        flights_dict['arrival_terminal'] = i['arrival'].get('terminal', None)
        flights_dict['local_arrival_sceduled_time'] = i['arrival'].get('scheduledTimeLocal', None)
        flights_dict['utc_arrival_sceduled_time'] = i['arrival'].get('scheduledTimeUtc', None)
        flights_dict['local_arrival_actual_time'] = i['arrival'].get('actualTimeLocal', None)
        flights_dict['utc_arrival_actual_time'] = i['arrival'].get('actualTimeUtc', None)  

        list_for_df.append(flights_dict)

  return pd.DataFrame(list_for_df)
```
